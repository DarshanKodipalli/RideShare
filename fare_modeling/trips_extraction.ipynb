```python
import dask.dataframe as dd
```

```python
df = dd.read_csv("/Volumes/MD/trips.csv", blocksize=128000000, dtype={'Trip Seconds': 'float64', 'Trips Pooled': 'int32'}, assume_missing=True)
```

```python
df.dropna()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
to_drop = ['Trip ID', 'Trip Start Timestamp', 'Trip End Timestamp',
       'Pickup Census Tract', 'Dropoff Census Tract',
       'Pickup Community Area', 'Dropoff Community Area', 'Fare', 'Tip',
       'Additional Charges', 'Pickup Centroid Latitude', 'Pickup Centroid Longitude',
       'Pickup Centroid Location', 'Dropoff Centroid Latitude',
       'Dropoff Centroid Longitude', 'Dropoff Centroid Location']
```

```python
df = df.drop(to_drop, axis=1)
```

```python
old_cols = df.columns.to_list()
```

```python
old_cols
```

```python
new_cols = ["DURATION", "DISTANCE", "FARE", "IS_POOLED", "TRIPS_POOLED"]
```

```python
df = df.rename(columns=dict(zip(old_cols, new_cols)))
```

```python
df
```

```python
df.compute()
```

```python
df.to_csv("trips_extracted.csv", index=False, single_file=True)
```
